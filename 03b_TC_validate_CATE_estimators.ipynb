```python
# Globals adapted to current dataset
from pathlib import Path

# Data and variable configuration
DATA_CSV = Path("data/analysis/analysis_df.csv")
outcome_name = "fausebal"  # set the target you want to validate
treatment_var = "message_fa"

# CV and evaluation settings
n_splits = 4
q_values = np.array([0.5, 0.4, 0.3, 0.2, 0.1])
n_bins = 5

# Base learners (aligned with tuning notebook)
base_learners = {
    "lasso": Lasso(),
    "logistic": LogisticRegression(solver="liblinear", max_iter=500),
    "rf": RandomForestRegressor(),
    "xgb": XGBRegressor(objective="reg:squarederror", n_jobs=1, tree_method="hist"),
}

# Derived later from data
ATE = None
dir_neg = None

# Save settings
SAVE_PATH = f"output/figures/{outcome_name}/"
savfig = True
```

```python
# Load and prepare current dataset (aligned with 02_tune_CATE_estimators)
import pandas as pd
import numpy as np
from pathlib import Path
from sklearn.impute import SimpleImputer

ANALYSIS_CSV = Path("data/analysis/analysis_df.csv")
if not ANALYSIS_CSV.exists():
    raise FileNotFoundError(f"Expected CSV not found at {ANALYSIS_CSV}. Run 00_master_analysis.ipynb or data prep to generate it.")

trainval_df = pd.read_csv(ANALYSIS_CSV)

# Build feature matrix similar to 02_tune
strat_vars = [c for c in trainval_df.columns if c.startswith("strat_")]
cat_candidates = ['reminder_freq', 'reminder_infreq', 'camp_short', 'htefa', 'htebal', 'awareness', 'creditcard']
low_card_cats = [c for c in cat_candidates if c in trainval_df.columns and trainval_df[c].nunique(dropna=True) <= 50]
cat_dummies = pd.get_dummies(trainval_df[low_card_cats], drop_first=True, dummy_na=False) if low_card_cats else pd.DataFrame(index=trainval_df.index)
X_strat = trainval_df[strat_vars].copy() if strat_vars else pd.DataFrame(index=trainval_df.index)
X_design = pd.concat([X_strat, cat_dummies], axis=1)
X_design = X_design.loc[:, ~X_design.columns.duplicated()].copy()

features = list(X_design.columns)
model_df = pd.concat([X_design, trainval_df[[treatment_var, outcome_name]]], axis=1)

# Impute features (median for numeric, most_frequent for non-numeric)
numeric_cols = model_df[features].select_dtypes(include=[np.number]).columns.tolist()
if numeric_cols:
    num_imp = SimpleImputer(strategy="median")
    X_numeric_imputed = pd.DataFrame(
        num_imp.fit_transform(model_df[numeric_cols]),
        columns=numeric_cols,
        index=model_df.index,
    )
else:
    X_numeric_imputed = pd.DataFrame(index=model_df.index)

non_numeric_cols = [c for c in features if c not in numeric_cols]
if non_numeric_cols:
    cat_imp = SimpleImputer(strategy="most_frequent")
    X_cats_imputed = pd.DataFrame(
        cat_imp.fit_transform(model_df[non_numeric_cols]),
        columns=non_numeric_cols,
        index=model_df.index,
    )
else:
    X_cats_imputed = pd.DataFrame(index=model_df.index)

X_design_imputed = pd.concat([X_numeric_imputed, X_cats_imputed], axis=1)[features]
model_df_imputed = pd.concat([X_design_imputed, model_df[[treatment_var, outcome_name]]], axis=1)

# Compute ATE from the full dataset and decide direction of interest
ATE = (model_df_imputed.loc[model_df_imputed[treatment_var] == 1, outcome_name].mean()
       - model_df_imputed.loc[model_df_imputed[treatment_var] == 0, outcome_name].mean())
dir_neg = (ATE < 0)

print(f"Prepared dataset for validation: {model_df_imputed.shape[0]} rows, {len(features)} features")
print(f"ATE={ATE:.4f}, dir_neg={dir_neg}")
```

```python
# Build fitted libraries using tuned params from output/params
import os
import joblib
from sklearn.model_selection import StratifiedKFold
from methods.cate_estimator_validation import make_estimator_library
from methods.data_processing import separate_vars as _separate

# Load tuned params saved by 02_tune
PARAMS_PATH = os.path.join('output', 'params', f'{outcome_name}_tuned_params.pkl')
if not os.path.exists(PARAMS_PATH):
    raise FileNotFoundError(f"Expected tuned params at {PARAMS_PATH}. Run 02_tune_CATE_estimators.ipynb.")

tuned_params = joblib.load(PARAMS_PATH)

# Create CV splitter
cv = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=405)

# Separate variables per current dataframe
X, t, y = _separate(model_df_imputed, outcome_name, treatment_var)

# Build library with fixed tuned params and fit on original (no perturb) fold definitions
library_none = make_estimator_library(X, t, y, cv, base_learners, tuned_params=tuned_params)
for estimator in library_none.values():
    estimator.fit()

# Optionally construct two additional CV perturbations
fitted_libraries = {"pert_none": library_none}
for i in range(2):
    cv_new = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=7*i)
    library_i = make_estimator_library(X, t, y, cv_new, base_learners, tuned_params=tuned_params)
    for estimator in library_i.values():
        estimator.fit()
    fitted_libraries[f"pert_cv_{i}"] = library_i

print("Fitted libraries:", list(fitted_libraries.keys()))
```

## 4. Stability of CATE estimators

### 4.1. Stability of quantile-based top subgroup

In this section, we examine the stability of the quantile-based subgroups by generalizing the notion of overlap.  
We consider the quantile-based top subgroup for q=0.2, and compute what percentage of its members are contained  
in the quantile-based top subgroup for other values of q, when the two subgroups are obtained from CATE estimators  
trained on different data splits.

```python
library_names = ["pert_none", "pert_cv_0", "pert_cv_1"]
library = {}
for estimator_name in fitted_libraries["pert_none"].keys():
    list_of_estimators = []
    for library_name in library_names:
        list_of_estimators.append(fitted_libraries[library_name][estimator_name])
    library[estimator_name] = combine_estimator_versions(list_of_estimators)
stability_scores = {}
q0 = 0.2
q_values_v2 = [0.2, 0.3, 0.4, 0.5]
for estimator_name, estimator in library.items():
    stability_scores[estimator_name] = get_top_subgroup_stability(estimator, q0, q_values_v2, dir_neg).round(2)
```

```python
same_q_scores = []
diff_q_scores = []
for scores in stability_scores.values():
    same_q_scores.append(scores.loc[0.2,"mean coverage"])
    diff_q_scores.append(scores.loc[0.3,"mean coverage"])
scores_df = pd.DataFrame({"estimator" : list(stability_scores.keys()),
                          "0.2" : same_q_scores,
                          "0.3" : diff_q_scores})
scores_df = scores_df.set_index("estimator")
scores_df.drop(index = ["r_lassoxgb"], inplace = True)
```

### 4.4. Pairwise correlation for a single estimator when trained on different splits

```python
marker_size = 20
alpha = 0.3

def make_model_CATE_pairplot(estimator, estimator_name = None):
    model_CATEs_df = pd.DataFrame({})
    for fold_no, result in estimator.results.items():
        model_CATEs_df[fold_no] = result.tau
    def corrfunc(x, y, **kwargs):
        r, _ = stats.pearsonr(x, y)
        ax = plt.gca()
        ax.annotate("r = {:.2f}".format(r),
                    xy = (.1, .9), xycoords = ax.transAxes,
                    fontsize = 20)
    g = sns.pairplot(model_CATEs_df, kind = "scatter", plot_kws = {"s" : marker_size,
                                                                   "alpha" : alpha})
    g = g.map_upper(corrfunc)
    g.fig.set_size_inches(15,15)
    title_string = "Pair plot for ITEs predicted by estimator when trained on 4 different folds"
    if estimator_name is not None:
        title_string += f": {estimator_name}"
    plt.suptitle(title_string, y = 1.03)
    plt.show()
```
